```python
new_col = []
for col_name in df1.columns:
    cols = col_name.lower().replace(' ', '_')
    new_col.append(cols)
df1.columns = new_col
df1.columns
```

```python
df1 = df1.rename(columns={"st":"state"})
df1.head()
```

```python
new_col2 = []
for col_name in df2.columns:
    cols = col_name.lower().replace(' ', '_')
    new_col2.append(cols)
df2.columns = new_col2
df2.columns
```

```python
df2 = df2.rename(columns={"st":"state"})
df2.head()
```

```python
new_col3 = []
for col_name in df3.columns:
    cols = col_name.lower().replace(' ', '_')
    new_col3.append(cols)
df3.columns = new_col3
df3.columns
```

```python
df3.head()
```

```python
data1 = df1[['customer', 'state', 'gender', 'policy_type', 'vehicle_class', 'education','customer_lifetime_value', 'income',
 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']]
data1.head()
```

```python
data2 = df2[['customer', 'state', 'gender', 'policy_type', 'vehicle_class', 'education','customer_lifetime_value', 'income',
 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']]
data2.head()
```

```python
data3 = df3[['customer', 'state', 'gender', 'policy_type', 'vehicle_class', 'education','customer_lifetime_value', 'income',
 'monthly_premium_auto', 'number_of_open_complaints', 'total_claim_amount']]
data3.head()
```

```python
data = pd.concat([df1, df2, df3])
data.head()
```

```python
data["customer_lifetime_value"]= data["customer_lifetime_value"].apply(lambda x: str(x))
data["customer_lifetime_value"]= data["customer_lifetime_value"].apply(lambda x: x.replace("%",""))
data["customer_lifetime_value"]= data["customer_lifetime_value"].fillna(0)
```
